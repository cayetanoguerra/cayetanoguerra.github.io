# **Práctica 1: Red Neuronal Convolutiva**
---

En esta práctica se va a desarrollar una red neuronal convolutiva para la clasificación de imágenes. Para ello, se debe utilizar el código de ejemplo inicial proporcionado en el material lectivo de la asignatura. Este código está escrito en PyTorch y será éste el framework que se utilice para esta práctica (no Keras ni Tensorflow).

A partir de este código inicial se deberán aplicar las transformaciones necesarias para adaptarlo a las condiciones del dataset elegido. Además, se requerirá realizar diferentes modelos modificando los hiperparámetros de la red (número de capas convolutivas y/o lineales, número de neuronas por capa, porcentaje de dropout, learninig rate, etc.) y estudiar y comparar los resultados obtenidos. El número de modelos con diferentes hiperparámetros dependerá del tiempo necesario para entrenarlos, como cifra orientativa podría ser un mínimo de cinco modelos.


## **Conjunto de datos**

El conjunto de datos puede provenir de Kaggle (https://www.kaggle.com/datasets) u otras fuentes de Internet, o ser realizado por los propios miembros del grupo. Debe constar de entre unas cuatro a diez categorías, contar con un mínimo de cien imágenes por categoría y estar dividido en conjunto de entrenamiento y validación. Estas cantidades son orientativas con el objetivo de que no consuma excesivos recursos computaciones y, por tanto, no se alargue excesivamente el tiempo de cómputo. En cualquier caso, el dataset debe contar con el visto bueno del profesor de prácticas.


### **Transfer Learning**

El transfer learning (aprendizaje por transferencia) es una técnica que aprovecha el conocimiento adquirido por un modelo previamente entrenado en una tarea para aplicarlo a una nueva tarea relacionada, reduciendo así el tiempo y los datos necesarios para el entrenamiento. Una vez realizada la red neuronal desde cero, se propone el uso de esta técnica y comprobar si ha habido mejora en los resultados.


### **Resultados**

Los resultados de los modelos deben incluir (por modelo):
•	Accuracy: Porcentaje de ejemplos correctamente clasificados.
•	Gráfica de la función de pérdida (loss) (entrenamiento y validación) y accuracy (entrenamiento y validación).
•	Matriz de confusión


### **Entregables**

Los resultados de esta práctica deberán presentarse como notebooks locales o en la nube (Google Colab o similares), comentando en celdas de texto los resultados obtenidos.

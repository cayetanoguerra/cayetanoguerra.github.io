# Probabilidad y verosimilitud

```python
import numpy as np

observations = np.random.choice(2, 1000, p=[0.2, 0.8] )

print(sum(observations))
print(observations)
```

```python
P = {0: 0.2, 1: 0.8}
Q = {0: 0.1, 1: 0.9}
R = {0: 0.7, 1: 0.3}

V_P = 1
V_Q = 1
V_R = 1

for o in observations:
    V_P *= P[o]
    V_Q *= Q[o]
    V_R *= R[o]
        
print(V_P)
print(V_Q)
print(V_R)
```

```python
V_P = 0
V_Q = 0
V_R = 0

for o in observations:
    V_P += np.log(P[o])
    V_Q += np.log(Q[o])
    V_R += np.log(R[o])
        
print(V_P)
print(V_Q)
print(V_R)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

def likelihood(p, obs):
    P = {0: p, 1: 1-p}
    V = 0
    for o in observations:
        V += np.log(P[o])
    return V
    
V = []
P = np.arange(0,1,0.01)

for p in P:
    V.append(likelihood(p, observations))
    
plt.plot(P,V)
plt.grid()
plt.show()
    
```
